```python
# importing some required libraries
from imageai.Detection import ObjectDetection
import os
import cv2
import time
# import sys
def forFrame(frame_number, output_array, output_count): ##defining a function for printing some output as per frame
    print("FOR FRAME " , frame_number)
    print("Output for each object : ")
    print("Object Name\t\tPercentage Probability\t\t Box Points")
    for obj in output_array:
        print(str(obj['name'])+"\t\t"+str(obj['percentage_probability'])+"\t\t"+str(obj['box_points']))
    print("Output count for unique objects : ")
    print("Object Name\t\tNumber of objects")
    for key in output_count.keys():
        print(str(key)+"\t\t"+str(output_count[key]))
    total_count=len(output_array)
    print("Total number of objects : ",total_count)
    print("------------END OF A FRAME --------------")

# def forSeconds(second_number, output_arrays, count_arrays, average_output_count):    ##defining a function for printing some output as per second
#     print("SECOND : ", second_number)
#     print("Array for the outputs of each frame ", output_arrays)
#     print("Array for output count for unique objects in each frame : ", count_arrays)
#     print("Output average count for unique objects in the last second: ", average_output_count)
#     print("------------END OF A SECOND --------------")

# def forMinute(minute_number, output_arrays, count_arrays, average_output_count):   ##defining a function for printing some output as per minute
#     print("MINUTE : ", minute_number)
#     print("Array for the outputs of each frame ", output_arrays)
#     print("Array for output count for unique objects in each frame : ", count_arrays)
#     print("Output average count for unique objects in the last minute: ", average_output_count)
#     print("------------END OF A MINUTE --------------")

execution_path = os.getcwd()         #assigning execution_path to the path of the current working directory
detector = ObjectDetection()        #Creating object of the ObjectDetection class
detector.setModelTypeAsYOLOv3()      #Setting type of the model which will be used for detecting object in the pictures or videos
detector.setModelPath(os.path.join(execution_path, "yolo.h5"))    #Setting the path of the model to the object of the ObjectDetection class 
detector.loadModel()       #Loading the model
# detector.loadModel(detection_speed="fast")
custom_objects = detector.CustomObjects(car=True, motorcycle=True, person=True)  #Customizing the object for detecting the particular objects

# limit=10
index=0   # declaring a variable which will contain the index of the camera
camera=1   #declaring a variable which will contain the number of camera will be used 
time_limit=20    #setting the max time limit which will be used for traffic management
tic=time.process_time()  # assigning the current system time to the tic variable 
cam = cv2.VideoCapture(index)   #creating the object of the VideoCapture class
while True:
    if cam.isOpened()==False:   #checking the camera is open or not, if not then
        cam.open(index)   # open the camera
    flag,frame=cam.read()    # capturing a frame from the camera here flag contains True if a frame is captured else False and
    if flag:                  #frame contains a frame
        #now calling the method for detecting object
        output_array = detector.detectCustomObjectsFromImage(custom_objects=custom_objects, input_image=frame, output_image_path=os.path.join(execution_path , "image3custom.jpg"),input_type="array", minimum_percentage_probability=30)
#         output_array=detector.detectObjectsFromImage(input_type="array",input_image=frame, output_image_path=os.path.join(execution_path, "traffic_detected2.jpg") )
    print("Camera : "+str(index))     #printing which camera is used for this frame
    print(output_array)        #printing the object name and  coordiante of each object detected in the frame
    toc=time.process_time()    
    if(((toc-tic)*1000000)>=time_limit or len(output_array)==0):   #chaecking if the time is over or no object detected in the
        cam.release()                                            #in the frame then release the current camera.
        index+=1
        index%=camera
        cam = cv2.VideoCapture(index)      #assigning the next camera to the cam variable
        tic=time.process_time()   #start the time for this camera
        
#     if index==0:
#         index=1
#     else :
#         index=0
#     print("Object Name\t\tPercentage Probability\t\t Box Points")
#     for obj in output_array:
#         print(str(obj['name'])+"\t\t"+str(obj['percentage_probability'])+"\t\t"+str(obj['box_points']))
#     print("----------------------------------------------------------------------")
```
